```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, new_name):
    self.__name = new_name

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"Student {self.__name} is already enrolled in {newCourse}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** .5


class Rectangle:
  def __init__(self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4
    self.width = self.width_()
    self.height = self.height_()

  def width_(self):
    return self.p1.distance(self.p2)

  def height_(self):
    return self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height

p1 = Point(0,0)
p2 = Point(5,0)
p3 = Point(0,5)
p4 = Point(5,5)
z = Rectangle(p1, p2, p3, p4)
z.area()
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__wage

  def set_wage(self, new_wage):
    if new_wage >= 0:
      self.__wage = new_wage
    else:
      print("Wage cannot be negative")

  def calculate_pay(self):
    return self.__wage * self.hours

  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Wage: {self.__wage}")
    print(f"Hours: {self.hours}")
    print(f"Pay: {self.calculate_pay()}")

  def __str__(self):
    return f"Employee: {self.name}\nWage: {self.__wage}\nHours: {self.hours}\nPay: {self.calculate_pay()}"


class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```
