```python
A = float(input("Enter the first number: "))
B = float(input("Enter the second number: "))
C = input("Enter an operator (+,-,*,/,//,%,**): ")
Addition = A + B
Subtraction = A - B
Multiplication = A * B
Division = A / B
Floor_Division = A // B
Modulus = A % B
Exponent = A ** B

if C == "+":
  print("Addition:", Addition)
elif C == "-":
  print("Subtraction:", Subtraction)
elif C == "*":
  print("Multiplication:", Multiplication)
elif C == "/":
    print("Division:", Division)
elif C == "//":
  print("Floor Division:", Floor_Division)
elif C == "%":
  print("Modulus:", Modulus)
elif C == "**":
  print("Exponent:", Exponent)
else:
  print("Invalid operator")
```

```python
import math

ticketPrice = 10
initAttendees = 20

print("Ads\tProfit")
for ad in range (0,201,25):
  revenue = (initAttendees + 2 *round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
```

```python
print("Hours\tFee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{hour}\t{fee}")
  hour += 1
```

```python
import math

ticketPrice = 10
initAttendees = 20

print("Ads\tProfit")

ad = 0
while ad < 201:
  revenue = (initAttendees + 2 *round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
  ad += 25
```
